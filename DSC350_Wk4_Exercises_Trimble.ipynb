```python
# Problem 3
import quandl

sunspots = quandl.get("SIDC/SUNSPOTS_A")  # Calls upon quandl to get the sunspots API. 
print("Head 2:\n", sunspots.head(2), "\n")  
print("Tail 2:\n", sunspots.tail(2), "\n")
last_date = sunspots.index[-1]
print("Last Value:\n",sunspots.loc[last_date], "\n")
print("Values sliced by data:\n", sunspots["20100101": "20201231"], "\n")
print("Observations greater than the mean:\n", sunspots[sunspots['Number of Observations'] > sunspots['Number of Observations'].mean()], "\n")
print("Sunspots greater than the mean:\n", sunspots[sunspots > sunspots.mean()], "\n")
```

```python
# Problem 4
print(f"Descriptive Statistics for Quandl Sunspots API\n")
sunspots = quandl.get("SIDC/SUNSPOTS_A")
print("Describe", sunspots.describe(),"\n")
print("Non NaN observations", sunspots.count(),"\n")
mad_values = sunspots.apply(lambda x: (x - x.mean()).abs().mean())
print("MAD", mad_values,"\n")
print("Mean", sunspots.mean(),"\n")
print("Median", sunspots.median(),"\n")
print("Max", sunspots.max(),"\n")
print("Min", sunspots.min(),"\n")
print("Mode", sunspots.mode(),"\n")
print("Standard Deviation", sunspots.std(),"\n")
print("Variance", sunspots.var(),"\n")
print("Skewness", sunspots.skew(),"\n")
```

Due to attribute errors "'DataFrame' object has no attribute 'mad'" and "'Series' object has no attribute 'mad'", MAD had to be calculated by using the apply and lambda functions.  
Lambda function calculates MAD by taking the absolute difference between each value and the mean, and then computes the mean of those absolute differences.  
According to research, mad() is not available in pandas (2.2.2). I reverted to the same version used in the course material, but the results were the same.    

```python
# Problem 5
import pandas as pd
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint
import numpy as np

seed(42)

# Creates our DataFrame and the values within each column of the DataFrame.
df = pd.DataFrame({'Weather' : ['cold', 'hot', 'cold', 'hot',
   'warm', 'cool', 'warm'],
   'Food' : ['soup', 'soup', 'icecream', 'sandwich',
   'icecream', 'sandwich', 'soup'],
   'Price' : 10 * rand(7), 'Number' : randint(1, 9)}) # Creates random numbers using NumPy.

print(df)
print()
weather_group = df.groupby('Weather')

i = 0  # Defines the instance to be used in the following lines of code.

for name, group in weather_group:  # Groups each name by the weather type.
   i = i + 1  # Adds one level to each group (Group 1, Group 2, etc.)
   print("Group", i, name)
   print(group)
print()
wf_group = df.groupby(['Weather', 'Food'])  # Creates Weather/Food group for aggregation.
print("Weather/Food Aggregated\n", wf_group.agg([np.mean, np.median]))  # Uses np.mean and np.median to aggregate the group.
```

```python
# Problem 7
# Creates DataFrames for the ingested CSVs.
dests = pd.read_csv('dest.csv')
tips = pd.read_csv('tips.csv')
# Merges fields based on a key column and excludes columns that are missing values.
print("Merge on key:\n", pd.merge(dests, tips, on='EmpNr'), "\n")
# Joins the dest and tips files.
print("File join()\n", dests.join(tips, lsuffix='Dest', rsuffix='Tips'), "\n")
```
